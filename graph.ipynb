```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
VGGNet_loss_hist = {
    "train":[],
    "val":[],
    "test":[],
}
VGGNet_acc_hist = {
    "train":[],
    "val":[],
    "test":[],
}
ResNet_loss_hist = {
    "train":[],
    "val":[],
    "test":[],
}
ResNet_acc_hist = {
    "train":[],
    "val":[],
    "test":[],
}
ResNext_loss_hist = {
    "train":[],
    "val":[],
    "test":[],
}
ResNext_acc_hist = {
    "train":[],
    "val":[],
    "test":[],
}

df_VGGNet = pd.read_excel('./graph/VGGNet.xlsx', header=0, index_col=0, names=None)
df_ResNet = pd.read_excel('./graph/ResNet.xlsx', header=0, index_col=0, names=None)
df_ResNext = pd.read_excel('./graph/ResNext.xlsx', header=0, index_col=0, names=None)

np_VGGNet = pd.DataFrame.to_numpy(df_VGGNet)
np_ResNet = pd.DataFrame.to_numpy(df_ResNet)
np_ResNext = pd.DataFrame.to_numpy(df_ResNext)

VGGNet_loss_hist["train"] = np_VGGNet[0,]
VGGNet_loss_hist["val"] = np_VGGNet[1,]
#VGGNet_loss_hist["test"] = np_VGGNet[2,]
ResNet_loss_hist["train"] = np_ResNet[0,]
ResNet_loss_hist["val"] = np_ResNet[1,]
ResNet_loss_hist["test"] = np_ResNet[2,]
ResNext_loss_hist["train"] = np_ResNext[0,]
ResNext_loss_hist["val"] = np_ResNext[1,]
ResNext_loss_hist["test"] = np_ResNext[2,]

VGGNet_acc_hist["train"] = np_VGGNet[2,]
VGGNet_acc_hist["val"] = np_VGGNet[3,]
#VGGNet_acc_hist["test"] = np_VGGNet[5,]
ResNet_acc_hist["train"] = np_ResNet[3,]
ResNet_acc_hist["val"] = np_ResNet[4,]
ResNet_acc_hist["test"] = np_ResNet[5,]
ResNext_acc_hist["train"] = np_ResNext[3,]
ResNext_acc_hist["val"] = np_ResNext[4,]
ResNext_acc_hist["test"] = np_ResNext[5,]

VGGNet_num_epochs = len(VGGNet_loss_hist["train"])
ResNet_num_epochs = len(ResNet_loss_hist["train"])
ResNext_num_epochs = len(ResNext_loss_hist["train"])
num_epochs = min(VGGNet_num_epochs, ResNet_num_epochs, ResNext_num_epochs)
```

```python
# plot Train loss progress
plt.title("Train Loss")
plt.plot(range(1,num_epochs+1),VGGNet_loss_hist["train"][0:num_epochs],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_loss_hist["train"][0:num_epochs],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_loss_hist["train"][0:num_epochs],label="ResNext")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
plt.show()

# plot Val loss progress
plt.title("Val Loss")
plt.plot(range(1,num_epochs+1),VGGNet_loss_hist["val"][0:num_epochs],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_loss_hist["val"][0:num_epochs],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_loss_hist["val"][0:num_epochs],label="ResNext")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
plt.show()

# plot test loss progress
'''
plt.title("Test Loss")
plt.plot(range(1,num_epochs+1),VGGNet_loss_hist["test"],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_loss_hist["test"],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_loss_hist["test"],label="ResNext")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.legend()
plt.show()
'''

# plot Train Accuracy progress
plt.title("Train Accuracy")
plt.plot(range(1,num_epochs+1),VGGNet_acc_hist["train"][0:num_epochs],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_acc_hist["train"][0:num_epochs],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_acc_hist["train"][0:num_epochs],label="ResNext")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
plt.show()

# plot Train Accuracy progress
plt.title("Val Accuracy")
plt.plot(range(1,num_epochs+1),VGGNet_acc_hist["val"][0:num_epochs],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_acc_hist["val"][0:num_epochs],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_acc_hist["val"][0:num_epochs],label="ResNext")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
plt.show()

# plot Test Accuracy progress
'''
plt.title("Test Accuracy")
plt.plot(range(1,num_epochs+1),VGGNet_acc_hist["test"],label="VGGNet")
plt.plot(range(1,num_epochs+1),ResNet_acc_hist["test"],label="ResNet")
plt.plot(range(1,num_epochs+1),ResNext_acc_hist["test"],label="ResNext")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
plt.show()
'''
```
